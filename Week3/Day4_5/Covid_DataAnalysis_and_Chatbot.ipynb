```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("country_vaccinations.csv")
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['Total_vaccinations(count)'] = df.groupby("country")['total_vaccinations'].tail(1)
df
```

```python
# top countries with most vaccines
df.groupby('country')['Total_vaccinations(count)'].mean().sort_values(ascending=False).head(10)
```

```python
#barplot visualization of top countries with most vaccinations
data = df.groupby('country')['Total_vaccinations(count)'].mean().sort_values(ascending=False).head(10)

sns.set_style("whitegrid")
ax = sns.barplot(y = data.index, x = data.values)
ax.set_xlabel('Total vaccinations(count)')
ax.set_title("top countries with most vaccinations")
plt.show()
```

```python
#Top countries with fully  vaccinated peoples
df['people_fully_vaccinated(count)'] = df.groupby('country')['people_fully_vaccinated'].tail(1)

data = df.groupby('country')['people_fully_vaccinated(count)'].mean().sort_values(ascending=False).head(15)

ax = sns.barplot(x=data.values, y=data.index)
ax.set_xlabel('people_fully_vaccinated(count)')
plt.show()
```

```python
vaccines = df.vaccines.unique()
list(vaccines)
```

```python
#most common vaccines
df.vaccines.value_counts(ascending=False).head(10)
```

```python
from wordcloud import WordCloud

plt.figure(figsize=(10, 10))
words = " ".join(df.vaccines)

image = WordCloud().generate_from_text(words)
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.countplot(data=df, y='vaccines')
```

```python
#daily vaccinations
data = df.groupby('date')['daily_vaccinations'].sum()

plt.figure(figsize=(15, 5))
sns.lineplot(data=data, x=data.index, y=data.values)
```

```python
#total vaccinations
plt.figure(figsize=(15,5))
sns.lineplot(data=df, x='date', y='total_vaccinations')
```

```python
#Countries with best daily average vaccinations
data = df.groupby('country')['daily_vaccinations'].mean().sort_values(ascending=False).head(15)

sns.barplot(y=data.index, x=data.values)
plt.show()
```

```python
# vaccinations in Pakistan
plt.figure(figsize=(15, 5))
sns.lineplot(x='date', y='people_fully_vaccinated', data=df[df.country=='Pakistan'])
```

```python
# top preferred vaccines in pakistan
df['vaccines'].value_counts(ascending=False).head(1).index.to_list()
```

```python
#COMPARING TOP 5 COUNTRIES WITH MOST VACCINATIONS
top_countries = list(df.groupby('country')['total_vaccinations'].mean().sort_values(ascending=False).head().index)
data = df.loc[df.country.apply(lambda x: x in top_countries)]

plt.figure(figsize=(15, 6))
sns.lineplot(x='date', y='total_vaccinations', data=data, hue='country', lw=2)
```

```python
#daily vaccination comparison
plt.figure(figsize=(15, 6))
sns.lineplot(x='date', y='daily_vaccinations', data=data, hue='country', lw=2)
```
