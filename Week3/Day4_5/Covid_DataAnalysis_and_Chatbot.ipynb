```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("country_vaccinations.csv")
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['Total_vaccinations(count)'] = df.groupby("country")['total_vaccinations'].tail(1)
df
```

```python
# top countries with most vaccines
df.groupby('country')['Total_vaccinations(count)'].mean().sort_values(ascending=False).head(10)
```

```python
#barplot visualization of top countries with most vaccinations
data = df.groupby('country')['Total_vaccinations(count)'].mean().sort_values(ascending=False).head(10)

sns.set_style("whitegrid")
ax = sns.barplot(y = data.index, x = data.values)
ax.set_xlabel('Total vaccinations(count)')
ax.set_title("top countries with most vaccinations")
plt.show()
```

```python
#Top countries with fully  vaccinated peoples
df['people_fully_vaccinated(count)'] = df.groupby('country')['people_fully_vaccinated'].tail(1)

data = df.groupby('country')['people_fully_vaccinated(count)'].mean().sort_values(ascending=False).head(15)

ax = sns.barplot(x=data.values, y=data.index)
ax.set_xlabel('people_fully_vaccinated(count)')
plt.show()
```

```python
vaccines = df.vaccines.unique()
list(vaccines)
```

```python
#most common vaccines
df.vaccines.value_counts(ascending=False).head(10)
```
