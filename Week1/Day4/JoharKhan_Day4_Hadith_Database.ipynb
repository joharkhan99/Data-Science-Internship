```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
# First we scrape Sunnah.com
url = "https://sunnah.com"

response = requests.get(url).text
doc = BeautifulSoup(response, "html.parser")

collections = doc.select(".collection_title")
output=[]
for c in collections:
    href = c.find("a").get('href')
    title = c.find("a").select(".english_collection_title")[0].text
    collection_url = url + href
    
    collection_response = requests.get(collection_url).text
    collection_doc = BeautifulSoup(collection_response, "html.parser")
    chapters = collection_doc.select(".book_title")

    for chap in chapters:
        chap_href = chap.find("a").get('href')
        chap_title = chap.find("a").select(".english_book_name")[0].text
        chap_url = url + chap_href
        print(chap_url)
        
        chap_response = requests.get(chap_url).text
        chap_doc = BeautifulSoup(chap_response, "html.parser")
        
        total_hadiths = chap_doc.select(".actualHadithContainer")
        
        for hadith in total_hadiths:        
            details = {}
            details['collection'] = title
            details['chapter'] = chap_title
            narrator_element = hadith.select(".hadith_narrated p")
            details['narrator'] = narrator_element[0].get_text() if len(narrator_element)>0 else "N/A"
            p_tags = hadith.select(".text_details p")
            details['description'] = " ".join([p.get_text() for p in p_tags])
            
            reference_element = hadith.select(".hadith_reference a")
            details['reference'] = reference_element[0].text if len(reference_element)>0 else "N/A"
            details['link'] = (url + hadith.select(".hadith_reference a")[0].get('href')) if len(reference_element)>0 else "N/A"

            output.append(details)

#             print(details)

# #         print(hadiths)

print("\n\n")
print(output)
```

```python
with open('hadith_data.csv', 'w', newline='', encoding='utf-8') as file:
    csv_writer = csv.writer(file)
    csv_writer.writerow(['Collection', 'Chapter', 'Narrator', 'Description', 'Reference', 'Link'])
    for item in output:
        csv_writer.writerow([item['collection'], item['chapter'], item['narrator'], item['description'], item['reference'], item['link']])
```
