### Read CSV

```python
import pandas as pd
# df = pd.read_csv("train.csv",nrows=20)
trainData = pd.read_csv("train.csv")
trainData
```

### Remove Duplicates

```python
trainData = trainData.drop_duplicates()
```

### Fill Missing Data

```python
mean = trainData['Age'].mean()
mean
trainData['Age'] = trainData['Age'].fillna(mean)
trainData
```

```python
# removing cabins column as there are alot missing and are not going to be used
trainData.drop(['Cabin'], axis=1, inplace=True)  #inplace changes the original dataset
trainData
```

### Calculate IQR (Inter Quartile Range)

```python
# 25%
Q1 = trainData['Fare'].quantile(0.25)
Q3 = trainData['Fare'].quantile(0.75)
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
lower_limit
```

```python
upper_limit
```

```python
# DROPPING OUTLIERS
outliers_low = (trainData['Fare']<lower_limit)
outliers_up = (trainData['Fare']>upper_limit)
```

```python
trainData[(outliers_low | outliers_up)]
```

### Drop Outliers

```python
trainData = trainData[~(outliers_low | outliers_up)]
trainData

# as you can see we have reduced 891 to 775 by removing outliers in Fare column
```
