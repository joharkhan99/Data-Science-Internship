## Relating the variables with scatterplots

```python
# relationship between total cases and recovered
sns.relplot(x="total_cases", y="discharged_healed",hue="discharged_healed", data=df)
```

```python
# relationship between total cases and deaths
sns.relplot(x="total_cases", y="deceased", data=df)
```

```python
# relationship between total cases and total cases tested
sns.relplot(x="total_cases", y="cases_tested",hue="cases_tested", data=df)
```

```python
# relationship between total_positives and total_hospitalised
sns.relplot(x="total_positives", y="hospitalized_with_symptoms", data=df)
```

```python
sns.pairplot(df)
```

```python
# so we can see the more home quarintine the more recovered cases
sns.relplot(x="discharged_healed", y="home_isolation",kind="line", data=df)
```

```python
sns.relplot(x="total_cases", y="total_hospitalised",kind="line", data=df)
```

```python
# representing categorical data...
sns.catplot(x="discharged_healed", y="total_cases", data=df)
```
